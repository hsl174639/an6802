```python
import pandas as pd
```

```python
df = pd.read_csv("FoodExpenditure.csv")
```

```python
df
```

```python
X = df.loc[:,["income"]]
Y = df.loc[:,["foodexp"]]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
pred = model.predict(X)
```

```python
rmse = mean_squared_error(Y,pred)**0.5
```

```python
rmse
```

```python
import joblib
```

```python
joblib.dump(model,"foodexp.pkl")
```

```python
model_saved = joblib.load("foodexp.pkl")
```
